# Present the Same Data Multiple Ways

```python
#import the dataset to the notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
df = pd.read_csv('avocado.csv')
df
```

## Plot one variable four different ways

```python
# 1
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.hist(df['AveragePrice'], bins=30)
plt.title('Distribution of Average Price of Avocados', fontsize=20)
plt.ylabel('Frequency of Type of Avocado', fontsize=16)
plt.xlabel('Average Price in USD', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:**
This histogram clearly shows the distribution of prices among avocados in the data set. 

**Cons:**
The plot does not show any more information about the data set.

```python
# 2
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.plot(df['AveragePrice'])
plt.title('Average Price of Avocados', fontsize=20)
plt.ylabel('Average Price in USD', fontsize=16)
plt.xlabel('Avocado', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** The line plot gives a sense of a general price increase as one moves down the index of avocados in the data set.

**Cons:** The data is scrambled together, and one cannot clearly see what the price of each avocado is.  

```python
# 3
plt.figure(figsize=(20, 10))
sns.set(style='darkgrid')
plt.boxplot(df['AveragePrice'])
plt.title('Average Price of Avocados', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** The plot is clean and clearly shows small variance among the price of avocados. 

**Cons:** The plot does not show anything else about the data set.

```python
# 4
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.scatter(df['AveragePrice'], df['Total Volume'])
plt.title('Relationship Between Average Price and Total Volume', fontsize=20)
plt.ylabel('Total Volume', fontsize=16)
plt.xlabel('Price in USD', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** The plot looks clean, and does a good job of showing little correlation between average price and total volume.

**Cons:** The dots at the bottom are jumbled together and could be confusing.

## Choose one continuous variable and one categorical variable, and plot them six different ways.

### Continuous: PLU 4225


```python
# 1
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.hist(df['4225'], bins=30)
plt.title('Distribution of Avocados with PLU 4225', fontsize=20)
plt.ylabel('Frequency of Avocados', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** The plot clearly shows that most avocados wit PLU 4046 have a high total sales volume. 

**Cons:** The distribution is difficult to read becasue teh data skews towards the left side of the plot.

```python
# 2
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.plot(df['4225'])
plt.title('Avocados with PLU 4225', fontsize=20)
plt.ylabel('Frequency of Avocados with PLU 4225', fontsize=16)
plt.xlabel('Total Sales Volume', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** The plot looks clean.

**Cons:** The data is jumbled, and it is difficult to see which sales volume values have the highest frequency. 

```python
#3
plt.figure(figsize=(20, 10))
sns.set()
plt.boxplot(df['4225'])
plt.ylim(0,600000)
plt.title('Variance in Avocados with PLU 4225', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** Very clean and easy to read. 

**Cons:** The data is skewed towards the lower end of the distribution, which makes the box plot look strange. 

```python
#4
import scipy.stats as stats
sns.set(style='whitegrid')
plt.figure(figsize=(20, 10))
stats.probplot(df['4225'], dist="norm", plot=plt)
plt.title('Avocados with PLU 4225 Q-Q plot', fontsize=20)
plt.ylabel('Ordered Values', fontsize=16)
plt.xlabel('Theoretical Quantities', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

**Pros:** Clean plot that shows that clearly shows that the total sales volume values of avocados with PLU 4225 do not follow a normal distribution.

**Cons:** Does not do much when not plotted against another variable from the data set.

```python
#5
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
sns.regplot(df['4225'], df['AveragePrice'])
plt.title('Relationship Between PLU 4225 and Average Price', fontsize=20)
plt.ylabel('Total Sales Volume', fontsize=16)
plt.xlabel('Average Price in USD', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
```

**Pros:** Clean plot that clearly shows the correlation between price and sales volume of PLU 4225.

**Cons:** Data jumbled on left side of chart. 

```python
#6
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
sns.residplot(df['4225'], df['AveragePrice'], lowess=True, color="b")
plt.title('Relationship Between PLU 4225 and Average Price', fontsize=20, fontstyle='oblique')
plt.ylabel('Total Sales Volume', fontsize=16)
plt.xlabel('Average Price in USD', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

**Pros:** Clean plot that shows the residuals of total sales volume against price.

**Cons:** Data jumbled on left side of chart. 

## Categorical: Region

```python
#1
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.bar(df['type'], df['AveragePrice'])
plt.title('Average Price by Type', fontsize=20, fontstyle='oblique')
plt.ylabel('Average Price in USD', fontsize=16)
plt.xlabel('Type', fontsize=16)
plt.tick_params(axis='both', which='both', labelsize=14)
plt.show()
```

**Pros:** Shows the average prices by type, very easy to read.

**Cons:** Too simple, not colorful or catching. 

```python
#2
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.plot(df['region'])
plt.title('Average Price by Region', fontsize=20, fontstyle='oblique')
plt.ylabel('Average Price in USD', fontsize=16)
plt.xlabel('Region', fontsize=16)
plt.tick_params(axis='x', which='both', labelsize=4)
plt.tick_params(axis='y', which='both', labelsize=14)
plt.show()
```

**Pros:** There are no pros here.

**Cons:** This chart makes no sense, the y labels are jumbled, and it just looks bad. 

```python
# 3
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
plt.barh(df['type'], df['AveragePrice'])
plt.title('Average Price by Type', fontsize=20, fontstyle='oblique')
plt.ylabel('Type', fontsize=16)
plt.xlabel('Average Price in USD', fontsize=16)
plt.tick_params(axis='both', which='both', labelsize=14)
plt.show()
```

**Pros:** Shows the average prices by type, very easy to read.

**Cons:** Too simple, not colorful or catching. 

```python
#4
sns.set()
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
ax = sns.boxplot(x="type", y="AveragePrice", data=df)
plt.xlabel('Type', fontsize=16)
plt.ylabel('Average Price in USD', fontsize=16)
plt.title('Variance in Average Price by Type', fontsize=20, fontstyle='oblique')
plt.tick_params(axis='both', which='both', labelsize=14)
```

**Pros:** Clean, clearly shows variance by categorical variable. 

**Cons:** Doesn't show other data from dataset. 

```python
#5
sns.set()
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
ax = sns.boxplot(x="type", y="4225", data=df)
plt.xlabel('Type', fontsize=16)
plt.ylabel('Total Sales Volume for Avocados with PLU 4225', fontsize=16)
plt.title('Variance in PLU 4225 by Type', fontsize=20, fontstyle='oblique')
plt.tick_params(axis='both', which='both', labelsize=14)
```

**Pros:** Clean, clearly shows variance by categorical variable. 

**Cons:** Data jumbled to bottom of chart. 

```python
# 6
sns.set()
plt.figure(figsize=(20, 10))
sns.set(style='whitegrid')
ax = sns.boxplot(x="type", y="Small Bags", data=df)
plt.xlabel('Type', fontsize=16)
plt.ylabel('Avocados Sold in Small Bags', fontsize=16)
plt.title('Variance in Avocados Sold in Small Bags', fontsize=20, fontstyle='oblique')
plt.tick_params(axis='both', which='both', labelsize=14)
```
